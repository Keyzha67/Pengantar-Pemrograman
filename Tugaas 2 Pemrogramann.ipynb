```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


def my_line(x, m, q):
    y = m * x + q
    return y


my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_majors', engine='openpyxl')

my_dataset1 = my_dataset[my_dataset.Epoch == 'one']
my_dataset2 = my_dataset[my_dataset.Epoch == 'two']

x = np.linspace(52.2, 62, 100)
y = my_line(x, m=0.3, q=-10.5)

fig, ax = plt.subplots()

ax.scatter(my_dataset1.SiO2, my_dataset1.CaO, marker= 'o', color="#E1F79B", edgecolor='#000000', label=r'$1^{st}$ Epoch')
ax.scatter(my_dataset2.SiO2, my_dataset2.K2O, marker= 'o', color="#505b398f", edgecolor='#000000', label=r'$2^{nd}$ Epoch')
ax.plot(x, y, color='#342a77')

ax.annotate(r'What is the 1$\sigma$ for this point?', xy=(52.6, 6.7), xytext=(47, 8.9), arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))
ax.text(52.4, 5.6, r'$ Na_2O = 0.3 \cdot SiO_2 -10.3$', dict(size=10, rotation=30))

ax.text(53.5, 5.1, r'$ \mu_{SiO_2} = \frac {a_{1}+a_{2}+\cdots +a_{n}}{n}$ = ' + '{:.1f} [wt.%]'.format(57.721),  dict(size=11.5))

ax.set_xlabel(r'SiO$_2$ [wt%]')
ax.set_ylabel(r'K$_2$O [wt%]')

ax.legend()
```
