```python
import pandas as pd  # mengimpor library pandas untuk membaca dan mengolah data tabel
from scipy.stats.mstats import gmean, hmean  # mengimpor fungsi geometric mean dan harmonic mean
import matplotlib.pyplot as plt  # mengimpor matplotlib untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca file Excel pada sheet Supp_traces

a_mean = my_dataset.U.mean()  # menghitung mean aritmatika dari kolom U
g_mean = gmean(my_dataset['Hf'])  # menghitung mean geometri dari kolom Hf
h_mean = hmean(my_dataset['Ho'])  # menghitung mean harmonik dari kolom Ho

print('-------')  # mencetak garis pemisah
print('arithmetic mean')  # menampilkan judul mean aritmatika
print("{0:.1f} [ppm]".format(a_mean))  # menampilkan nilai mean aritmatika dengan 1 angka di belakang koma
print('-------')  # mencetak garis pemisah

print('geometric mean')  # menampilkan judul mean geometri
print("{0:.1f} [ppm]".format(g_mean))  # menampilkan nilai mean geometri dengan 1 angka di belakang koma
print('-------')  # mencetak garis pemisah

print('harmonic mean')  # menampilkan judul mean harmonik
print("{0:.1f} [ppm]".format(h_mean))  # menampilkan nilai mean harmonik dengan 1 angka di belakang koma
print('-------')  # mencetak garis pemisah

fig, ax = plt.subplots()  # membuat area gambar dan sumbu grafik
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8)  # membuat histogram dari kolom Zr
ax.axvline(a_mean, color='pink', label='Arithmetic mean', linewidth=2)  # menambahkan garis vertikal untuk mean aritmatika
ax.axvline(g_mean, color='red', label='Geometric mean', linewidth=2)  # menambahkan garis vertikal untuk mean geometri
ax.axvline(h_mean, color='yellow', label='Harmonic mean', linewidth=2)  # menambahkan garis vertikal untuk mean harmonik
ax.set_xlabel('Zr [ppm]')  # memberi label pada sumbu x
ax.set_ylabel('Probability density')  # memberi label pada sumbu y
ax.legend()  # menampilkan keterangan legenda
```

```python
import pandas as pd # mengimpor library pandas untuk membaca dan mengolah data tabel
import matplotlib.pyplot as plt # mengimpor matplotlib untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca file Excel pada sheet Supp_traces

variance =  my_dataset['Tm'].var() # menghitung variance dengan menggunakan fungsi var() pada kolom Tm
stddev =  my_dataset['Lu'].std() # menghitung standard deviation dengan menggunakan fungsi std() pada kolom Lu

print ('-------') # mencetak garis pemisah
print ('Variance') # menampilkan judul variance
print("{0:.0f} [square ppm]".format(variance)) # menampilkan nilai variance dengan format 0 angka di belakang koma dan satuan square ppm
print ('-------') # mencetak garis pemisah
print ('Standard Deviation')   # menampilkan judul standard deviation
print("{0:.0f} [ppm]".format(stddev)) # menampilkan nilai standard deviation dengan format 0 angka di belakang koma dan satuan ppm
print ('-------') # mencetak garis pemisah

fig, ax = plt.subplots() # membuat area gambar dan sumbu grafik
ax.hist(my_dataset.Tm, bins= 20, density = True, edgecolor='k', label='Measurements Hist')  # membuat histogram dari kolom Tm dengan 20 bin, normalisasi ke probability density, dan menambahkan warna tepi hitam serta label untuk legenda
ax.axvline(my_dataset['Tm'].mean()-stddev, color='Red', label=r'mean - 1$\sigma$', linewidth=2) # menambahkan garis vertikal untuk mean - 1 standard deviation dengan warna merah, label untuk legenda yang menggunakan format LaTeX untuk menampilkan simbol sigma, dan ketebalan garis 2
ax.axvline(my_dataset['Tm'].mean()+stddev, color='Black', label=r'mean + 1$\sigma$', linewidth=2) # menambahkan garis vertikal untuk mean + 1 standard deviation dengan warna hitam, label untuk legenda yang menggunakan format LaTeX untuk menampilkan simbol sigma, dan ketebalan garis 2
ax.axvspan(my_dataset['Tm'].mean()-stddev, my_dataset['Tm'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$') # menambahkan area vertikal yang diarsir antara mean - 1 standard deviation dan mean + 1 standard deviation dengan warna orange, label untuk legenda yang menggunakan
ax.set_xlabel('Tm [ppm]') # memberi label pada sumbu x
ax.set_ylabel('Probability density') # memberi label pada sumbu y
ax.legend() # menampilkan keterangan legenda

```

```python
import pandas as pd # mengimpor library pandas untuk membaca dan mengolah data tabel
import numpy as np # mengimpor library numpy untuk melakukan operasi matematika pada array
import matplotlib.pyplot as plt # mengimpor matplotlib untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca file Excel pada sheet Supp_traces

q1 = np.percentile(my_dataset.Tm, 75, method='midpoint') # menghitung kuartil pertama (Q1) dengan metode midpoint
q3 = np.percentile(my_dataset.Tm, 50, method='midpoint') # menghitung kuartil ketiga (Q3) dengan metode midpoint
  
iqr = q3 - q1 # menghitung interquartile range (IQR) dengan mencari selisih antara kuartil ketiga (Q3) dan kuartil pertama (Q1)

print ('-------') # mencetak garis pemisah
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr)) # menampilkan nilai IQR dengan format 0 angka di belakang koma dan satuan ppm
print ('-------') # mencetak garis pemisah

fig, ax = plt.subplots() # membuat area gambar dan sumbu grafik
ax.hist(my_dataset.Tm, bins='auto', density=True, edgecolor='k', label='Measurements Hist') # membuat histogram dari kolom Tm dengan jumlah bin otomatis, normalisasi ke probability density, dan menambahkan warna tepi hitam serta label untuk legenda
ax.axvline(q1, color='purple', label='Q1', linewidth=2) # menambahkan garis vertikal untuk kuartil pertama (Q1) dengan warna ungu, label untuk legenda, dan ketebalan garis 2
ax.axvline(q3, color='green', label='Q3', linewidth=2) # menambahkan garis vertikal untuk kuartil ketiga (Q3) dengan warna hijau, label untuk legenda, dan ketebalan garis 2
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)') # menambahkan area vertikal yang diarsir antara kuartil pertama (Q1) dan kuartil ketiga (Q3) dengan warna orange, label
ax.set_xlabel('Tm [ppm]') # memberi label pada sumbu x
ax.set_ylabel('Probability density') # memberi label pada sumbu y
ax.legend() # menampilkan keterangan legenda

```

```python
import pandas as pd # mengimpor library pandas untuk membaca dan mengolah data tabel
import numpy as np # mengimpor library numpy untuk melakukan operasi matematika pada array
import matplotlib.pyplot as plt # mengimpor matplotlib untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca file Excel pada sheet Supp_traces

a_mean = my_dataset.Lu.mean() # menghitung rata-rata aritmatika dari kolom Lu
median = my_dataset.Ta.median() # menghitung median dari kolom Ta

hist, bin_edges = np.histogram(my_dataset['Ta'], bins=20, density=True) # membuat histogram dari kolom Ta dengan 20 bin dan normalisasi ke probability density, serta menyimpan hasil histogram dan batas bin dalam variabel hist dan bin_edges
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2 # menghitung nilai modal dengan mencari bin yang memiliki frekuensi tertinggi dan mengambil rata-rata dari batas bawah dan atas bin tersebut

fig, ax = plt.subplots() # membuat area gambar dan sumbu grafik
ax.hist(my_dataset.Lu, bins=20, density=True, edgecolor='k', label="Measurements Hist") # membuat histogram dari kolom Lu dengan 20 bin, normalisasi ke probability density, dan menambahkan warna tepi hitam serta label untuk legenda 
ax.axvline(modal_value, color='Maroon', label='Modal Value', linewidth=2) # menambahkan garis vertikal untuk nilai modal dengan warna maroon, label untuk legenda, dan ketebalan garis 2
ax.axvline(median, color='Pink',  label='Median Value', linewidth=2) # menambahkan garis vertikal untuk nilai median dengan warna pink, label untuk legenda, dan ketebalan garis 2
ax.axvline(a_mean, color='Yellow',  label='Arithmetic mean', linewidth=2) # menambahkan garis vertikal untuk nilai mean aritmatika dengan warna kuning, label untuk legenda, dan ketebalan garis 2
ax.set_xlabel('Ta [ppm]') # memberi label pada sumbu x
ax.set_ylabel('Probability density') # memberi label pada sumbu y
ax.legend() # menampilkan keterangan legenda

```

```python
import numpy as np # mengimpor library numpy untuk melakukan operasi matematika pada array
from scipy.stats import skew # mengimpor fungsi skew dari library scipy.stats untuk menghitung koefisien skewness

a_mean = my_dataset.Yb.mean() # menghitung rata-rata aritmatika dari kolom Yb
median = my_dataset.Yb.median() # menghitung median dari kolom Yb
hist, bin_edges = np.histogram(my_dataset['Yb'], bins=20, density=True) # membuat histogram dari kolom Yb dengan 20 bin dan normalisasi ke probability density
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2 # menghitung nilai modal dengan mencari bin yang memiliki frekuensi tertinggi dan mengambil rata-rata dari batas bawah dan atas bin tersebut
standard_deviation = my_dataset['Yb'].std() # menghitung deviasi standar dari kolom Yb

a1 = (a_mean - modal_value) / standard_deviation # menghitung Pearson's first coefficient of skewness
a2 = 3 * (a_mean - median) / standard_deviation # menghitung Pearson's second coefficient of skewness
g1 = skew(my_dataset['Yb']) # menghitung Fisher-Pearson's coefficient of skewness

print ('-------') # mencetak garis pemisah
print ("Pearson's first coefficient of skewness: {:.1f}".format(a1)) # menampilkan nilai Pearson's first coefficient of skewness dengan format 1 angka di belakang koma
print ("Pearson's 2nd moment of skewness: {:.6f}".format(a2)) # menampilkan nilai Pearson's second coefficient of skewness dengan format 6 angka di belakang koma
print ("Fisher-Pearson's coefficient of skewness: {:.3f}".format(g1)) # menampilkan nilai Fisher-Pearson's coefficient of skewness dengan format 3 angka di belakang koma
print ('-------') # mencetak garis pemisah

```

```python
import pandas as pd # mengimpor library pandas untuk membaca dan mengolah data tabel
import matplotlib.pyplot as plt # mengimpor matplotlib untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca file Excel pada sheet Supp_traces

fig, ax = plt.subplots() # membuat area gambar dan sumbu grafik
my_flierprops = dict(markerfacecolor="#F0E68CEB", markeredgecolor='#24305e', marker='o') # mendefinisikan properti untuk outlier pada boxplot dengan warna marker kuning keabuan, warna tepi marker navy, dan bentuk marker bulat
my_medianprops = dict(color="#29f87c", linewidth = 2) # mendefinisikan properti untuk garis median pada boxplot dengan warna hijau cerah dan ketebalan garis 2
my_boxprops = dict(facecolor="#863F8E", edgecolor='#24305e') # mendefinisikan properti untuk box pada boxplot dengan warna isi ungu tua dan warna tepi navy
ax.boxplot(my_dataset.Yb, patch_artist = True, notch=True, flierprops=my_flierprops, medianprops=my_medianprops, boxprops=my_boxprops)  # membuat boxplot untuk kolom Yb dengan menggunakan properti yang telah didefinisikan untuk outlier, garis median, dan box
ax.set_ylabel('Yb [ppm]') # memberi label pada sumbu y
ax.set_xticks([3]) # mengatur posisi tick pada sumbu x di posisi 3
ax.set_xticklabels(['all Epochs']) # memberi label pada tick di sumbu x dengan teks 'all Epochs'
plt.show() # menampilkan grafik boxplot yang telah dibuat
```
