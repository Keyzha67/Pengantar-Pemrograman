```python
import numpy as np # mengimpor library numpy untuk melakukan operasi matematika dan manipulasi array

def integrate_rec(f, a, b, n): # mendefinisikan fungsi integrate_rec yang menerima empat argumen: f (fungsi yang akan diintegrasikan), a (batas bawah integrasi), b (batas atas integrasi), dan n (jumlah subinterval yang digunakan dalam metode rectangle)
    # Implementation of the rectanlge method 
    h = (b-a)/n # menghitung lebar setiap subinterval dengan membagi selisih antara batas atas dan batas bawah dengan jumlah subinterval n
    x = np.linspace(a, b, n+1) # membuat array x yang berisi titik-titik pembagi interval dari a hingga b dengan jumlah n+1 titik (termasuk batas bawah dan batas atas)
    i=0 # inisialisasi variabel i untuk iterasi
    area=0 # inisialisasi variabel area untuk menyimpan hasil integrasi
    while i<n: # melakukan iterasi sebanyak n kali untuk menghitung luas area di bawah kurva menggunakan metode rectangle
        sup_rect =  f(x[i])*h # menghitung luas rectangle dengan menggunakan nilai fungsi f pada titik x[i] sebagai tinggi dan h sebagai lebar
        area += sup_rect # menambahkan luas rectangle yang dihitung ke variabel area untuk mendapatkan total luas di bawah kurva
        i += 1  # increment i untuk melanjutkan ke titik berikutnya
    return area # mengembalikan hasil integrasi yang telah dihitung menggunakan metode rectangle
'''
We test the Rectangle method on the sine funcion were the definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_7 = integrate_rec(np.sin, 0, np.pi/2, 7) # menghitung pendekatan integral dari fungsi sinus pada interval [0, π/2] menggunakan metode rectangle dengan jumlah subinterval n=7
sup_14 = integrate_rec(np.sin, 0, np.pi/2, 14) # menghitung pendekatan integral dari fungsi sinus pada interval [0, π/2] menggunakan metode rectangle dengan jumlah subinterval n=14
sup_21 = integrate_rec(np.sin, 0, np.pi/2, 21) # menghitung pendekatan integral dari fungsi sinus pada interval [0, π/2] menggunakan metode rectangle dengan jumlah subinterval n=21

print('Using n=7, the rectangle method returns a value of {:.2f}'.format(sup_7)) # mencetak hasil integrasi yang dihitung dengan menggunakan metode rectangle untuk n=7, dengan format dua angka di belakang koma
print('Using n=14, the rectangle method returns a value of {:.2f}'.format(sup_14)) # mencetak hasil integrasi yang dihitung dengan menggunakan metode rectangle untuk n=14, dengan format dua angka di belakang koma
print('Using n=21, the rectangle method returns a value of {:.2f}'.format(sup_21)) # mencetak hasil integrasi yang dihitung dengan menggunakan metode rectangle untuk n=21, dengan format dua angka di belakang koma
     
```

```python
import numpy as np # mengimpor library numpy untuk melakukan operasi matematika dan manipulasi array

def integrate_trap(f, a, b, n): # mendefinisikan fungsi integrate_trap yang menerima empat argumen: f (fungsi yang akan diintegrasikan), a (batas bawah integrasi), b (batas atas integrasi), dan n (jumlah subinterval yang digunakan dalam metode trapezoidal)
    # Implementation of the trapezoidal rule
    h = (b-a)/n # menghitung lebar setiap subinterval dengan membagi selisih antara batas atas dan batas bawah dengan jumlah subinterval n
    x = np.linspace(a, b, n+1) # membuat array x yang berisi titik-titik pembagi interval dari a hingga b dengan jumlah n+1 titik (termasuk batas bawah dan batas atas)
    i=1 # inisialisasi variabel i untuk iterasi, dimulai dari 1 karena titik pertama (x[0]) dan titik terakhir (x[n]) akan dihitung secara terpisah dalam rumus trapezoidal
    area = h*(f(x[0]) + f(x[n]))/2 # menghitung luas area di bawah kurva menggunakan rumus trapezoidal, dengan menghitung nilai fungsi f pada titik pertama (x[0]) dan titik terakhir (x[n]), kemudian menjumlahkannya dan membaginya dengan 2, lalu mengalikan dengan lebar subinterval h
    while i<n: # melakukan iterasi sebanyak n-1 kali untuk menghitung luas area di bawah kurva menggunakan metode trapezoidal, dimulai dari titik kedua (x[1]) hingga titik sebelum terakhir (x[n-1])
        sup_rect = f(x[i])*h # menghitung luas rectangle dengan menggunakan nilai fungsi f pada titik x[i] sebagai tinggi dan h sebagai lebar, yang akan digunakan untuk menambahkan ke area total
        area += sup_rect  # menambahkan luas rectangle yang dihitung ke variabel area untuk mendapatkan total luas di bawah kurva
        i += 1 # increment i untuk melanjutkan ke titik berikutnya
    return area # mengembalikan hasil integrasi yang telah dihitung menggunakan metode trapezoidal

'''
We test the trapezoidal rule on the known sine funcion were the  
definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_3 = integrate_trap(np.sin, 0, np.pi/2, 3) # menghitung pendekatan integral dari fungsi sinus pada interval [0, π/2] menggunakan metode trapezoidal dengan jumlah subinterval n=3
sup_6 = integrate_trap(np.sin, 0, np.pi/2, 6) # menghitung pendekatan integral dari fungsi sinus pada interval [0, π/2] menggunakan metode trapezoidal dengan jumlah subinterval n=6

print('Using n=3, the trapezoidal rule returns a value of {:.2f}'.format(sup_3)) # mencetak hasil integrasi yang dihitung dengan menggunakan metode trapezoidal untuk n=3, dengan format dua angka di belakang koma
print('Using n=6, the trapezoidal rule returns a value of {:.2f}'.format(sup_6)) # mencetak hasil integrasi yang dihitung dengan menggunakan metode trapezoidal untuk n=6, dengan format dua angka di belakang koma
   
```

```python
import numpy as np # mengimpor library numpy untuk operasi matematika dan pembuatan array
from scipy.integrate import trapezoid # mengimpor fungsi trapezoid dari modul integrate di scipy untuk melakukan integrasi numerik menggunakan metode trapezoidal
import matplotlib.pyplot as plt # mengimpor library matplotlib.pyplot untuk membuat visualisasi grafik

r = np.linspace(1,6400,6400) # membuat array r yang berisi 6400 titik yang terdistribusi secara merata dari 1 hingga 6400, yang mungkin mewakili jarak dari pusat bumi dalam kilometer

def density(): # mendefinisikan fungsi density yang akan menghitung dan mengembalikan array densitas (ro) berdasarkan model struktur internal bumi
    ro_inner_core = np.linspace(12200, 9800, 2259) # membuat array ro_inner_core yang berisi 2259 titik yang terdistribusi secara merata dari 12200 hingga 9800, yang mungkin mewakili densitas pada inti dalam bumi
    ro_outer_core = np.linspace(13100, 12800, 1220) # membuat array ro_outer_core yang berisi 1220 titik yang terdistribusi secara merata dari 13100 hingga 12800, yang mungkin mewakili densitas pada inti luar bumi
    ro_lower_mantle = np.linspace(4130,3312,2171) # membuat array ro_lower_mantle yang berisi 2171 titik yang terdistribusi secara merata dari 4130 hingga 3312, yang mungkin mewakili densitas pada mantel bawah bumi
    ro_upper_mantle = np.linspace(6100,4400,720) # membuat array ro_upper_mantle yang berisi 720 titik yang terdistribusi secara merata dari 6100 hingga 4400, yang mungkin mewakili densitas pada mantel atas bumi
    ro_crust = np.linspace(3100,2700,30) # membuat array ro_crust yang berisi 30 titik yang terdistribusi secara merata dari 3100 hingga 2700, yang mungkin mewakili densitas pada kerak bumi
   
    ro_final = np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust)) # menggabungkan semua array densitas (ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust) menjadi satu array ro_final menggunakan fungsi concatenate dari numpy
    
    return ro_final # mengembalikan array ro_final yang merupakan hasil penggabungan dari semua array densitas yang mewakili berbagai lapisan struktur internal bumi

ro = density() # memanggil fungsi density untuk mendapatkan array ro yang berisi nilai-nilai densitas berdasarkan model struktur internal bumi, dan menyimpan hasilnya dalam variabel ro

fig, ax = plt.subplots() # membuat sebuah figure dan axes menggunakan fungsi subplots dari matplotlib.pyplot, yang akan digunakan untuk membuat grafik densitas terhadap jarak dari pusat bumi
ax.plot(r,ro, label="Densities on the Earth's Interior") # membuat grafik dengan menggunakan fungsi plot dari axes (ax) untuk memplot array r (jarak dari pusat bumi) pada sumbu x dan array ro (densitas) pada sumbu y, serta memberikan label "Densities on the Earth's Interior" untuk legenda
ax.set_ylabel(r"Density [Kg/m$^3$]") # mengatur label untuk sumbu y dengan menggunakan fungsi set_ylabel dari axes (ax), dan menggunakan raw
ax.set_xlabel("Distance from the Earth center r [km]") # mengatur label untuk sumbu x dengan menggunakan fungsi set_xlabel dari axes (ax), dan memberikan deskripsi "Distance from the Earth center r [km]" untuk menjelaskan bahwa sumbu x mewakili jarak dari pusat bumi dalam kilometer
ax.legend() # menampilkan legenda pada grafik dengan menggunakan fungsi legend dari axes (ax), yang akan menampilkan label yang telah ditentukan sebelumnya untuk grafik
ax.grid() # menambahkan grid pada grafik dengan menggunakan fungsi grid dari axes (ax) untuk memudahkan pembacaan nilai pada grafik
```

```python
def gravity(r):  # mendefinisikan fungsi untuk menghitung percepatan gravitasi berdasarkan radius r
    
    g = np.zeros(len(r))  # membuat array nol untuk menyimpan nilai gravitasi
    Gr = 6.67408e-11  # konstanta gravitasi (G)
    r = np.linspace(1,6323,6300) # membuat array r yang berisi 6300 titik yang terdistribusi secara merata dari 1 hingga 6323, yang mungkin mewakili jarak dari pusat bumi dalam kilometer
    
    for i in range(1,len(r)):  # melakukan iterasi mulai dari indeks ke-1 hingga akhir
        
        r1 = r[0:i]  # mengambil sebagian data radius dari awal sampai indeks i
        ro1  = ro[0:i]  # mengambil sebagian data densitas (rho)
        r2 = r1[i-1]  # mengambil nilai radius terakhir pada subset
    
        y = ro1*r1**2  # menghitung nilai fungsi integrasi
        y_int = trapezoid(y,r1)  # menghitung integral menggunakan metode trapezoid
        
        g1 = ((4 * np.pi*Gr)/(r2**2)) * y_int  # menghitung nilai gravitasi
        g[i] = g1  # menyimpan hasil ke array g
           
    return g  # mengembalikan array gravitasi

g = gravity(r)  # memanggil fungsi gravity dengan input r

fig, ax = plt.subplots()  # membuat figure dan axis untuk plot
ax.plot(r,g)  # menampilkan grafik hubungan r terhadap g
ax.grid()  # menampilkan grid pada grafik
ax.set_ylabel(r'g [m/s$^2]$')  # memberi label sumbu y
ax.set_xlabel('Distance from the Earth center r [km]')  # memberi label sumbu x
```

```python
def pressure(r, ro, g): # mendefinisikan fungsi untuk menghitung tekanan berdasarkan radius r, densitas ro, dan percepatan gravitasi g
    
    p = np.zeros(len(r)) # membuat array nol untuk menyimpan nilai tekanan
    r = np.linspace(1,6323,6300) # membuat array r yang berisi 6300 titik yang terdistribusi secara merata dari 1 hingga 6323, yang mungkin mewakili jarak dari pusat bumi dalam kilometer
    
    for i in range(0,len(r)): # melakukan iterasi mulai dari indeks ke-0 hingga akhir
        r1 = r[i:len(r)] # mengambil sebagian data radius dari indeks i sampai akhir
        ro1 = ro[i:len(r)] # mengambil sebagian data densitas (rho) dari indeks i sampai akhir
        g1 = g[i:len(r)] # mengambil sebagian data percepatan gravitasi dari indeks i sampai akhir
        y = ro1*g1 # menghitung nilai fungsi integrasi untuk tekanan dengan mengalikan densitas (ro1) dengan percepatan gravitasi (g1)
        p1 = trapezoid(y,r1) # menghitung integral menggunakan metode trapezoid untuk mendapatkan tekanan pada radius r[i]
        p[i] = p1 # menyimpan hasil tekanan yang dihitung ke array p pada indeks i
    return p # mengembalikan array tekanan yang telah dihitung berdasarkan radius r, densitas ro, dan percepatan gravitasi g
    
p = pressure(r,ro,g)/1e9 # menghitung tekanan dengan memanggil fungsi pressure dan membagi hasilnya dengan 1e9 untuk mengkonversi dari Pascal ke GigaPascal (GPa)
z = np.linspace(1, 6323, 6400) # membuat array z yang berisi 6400 titik yang terdistribusi secara merata dari 1 hingga 6323, yang mungkin mewakili kedalaman dari permukaan bumi dalam kilometer

fig, ax = plt.subplots()  # membuat figure dan axis untuk plot
ax.plot(z, p[:len(z)]) # menampilkan grafik hubungan z terhadap p, dengan memotong array p agar sesuai dengan panjang array z
ax.grid() # menampilkan grid pada grafik
ax.set_ylabel('P [GPa]') # memberi label sumbu y dengan satuan GigaPascal (GPa)
ax.set_xlabel('Depth z from the Earth Surface [km]') # memberi label sumbu x dengan deskripsi "Depth z from the Earth Surface [km]" untuk menjelaskan bahwa sumbu x mewakili kedalaman dari permukaan bumi dalam kilometer
```
